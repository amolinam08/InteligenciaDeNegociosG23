###Datos con Asma 2017

```python
# Verificar la unicidad de los registros
duplicates = df_Asma2017.duplicated()
if duplicates.any():
    print("Se encontraron registros duplicados. Eliminando duplicados...")
    df_Asma2017 = df_Asma2017[~duplicates]

# Verificar la completitud de los datos
missing_values = df_Asma2017.isnull().sum()
if missing_values.any():
    print("Se encontraron valores faltantes. Realizando imputación...")
    # Realizar la imputación de valores faltantes según la estrategia adecuada

# Realizar validaciones adicionales según las necesidades del conjunto de datos

# Eliminar filas con valores nulos o "NA" en las columnas NPCEP4, NVCBP11A y NOMBRE_LOCALIDAD
df_Asma2017 = df_Asma2017.dropna(subset=['NPCEP4', 'NVCBP11A', 'NOMBRE_LOCALIDAD'])

# Renombrar columnas
df_Asma2017 = df_Asma2017.rename(columns={'NPCEP4': 'Edad', 'NVCBP11A': 'Estrato', 'NHCMP12P': 'Tiene_perro'})

# Guardar los datos limpios en un nuevo archivo CSV
df_Asma2017.to_csv('sample_data/datos_asma2017_clean.csv', index=False)
```

###Datos con Asma 2021

```python

# Verificar la unicidad de los registros
duplicates = df_Asma2021.duplicated()
if duplicates.any():
    print("Se encontraron registros duplicados. Eliminando duplicados...")
    df_Asma2021 = df_Asma2021[~duplicates]

# Verificar la completitud de los datos
missing_values = df_Asma2021.isnull().sum()
if missing_values.any():
    print("Se encontraron valores faltantes. Realizando imputación...")
    # Realizar la imputación de valores faltantes según la estrategia adecuada

# Realizar validaciones adicionales según las necesidades del conjunto de datos

# Eliminar filas con valores nulos o "NA" en las columnas especificadas
columns_to_dropna = ['NVCBP11A', 'NHCCP47A', 'NHCCP47A1', 'NHCCP47B', 'NHCCP47B1', 'NPCEP_5']
df_Asma2021 = df_Asma2021.dropna(subset=columns_to_dropna)

# Renombrar columnas
df_Asma2021 = df_Asma2021.rename(columns={'NVCBP11A': 'Estrato',
                                          'NHCCP47A': 'Tiene_perro',
                                          'NHCCP47A1': 'Num_perros',
                                          'NHCCP47B': 'Tiene_gatos',
                                          'NHCCP47B1': 'Num_gatos',
                                          'NPCEP_5': 'Edad'})

# Guardar los datos limpios en un nuevo archivo CSV
df_Asma2021.to_csv('sample_data/datos_asma2021_clean.csv', index=False)
```
